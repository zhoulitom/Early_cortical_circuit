```python
#get started here! 10-29, 2019
#get started here, 05-11-2020
#20200731
#08252020, rotate the heatmap
#select synaptic markers and output a raw fig, and a .CSV file

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

gene_name = pd.read_table('GRCh38_genes_ENS.txt', header=0 ) #header, index_col
gene_name.rename(columns ={'Ensembl Gene ID':'gene','Associated Gene Name':'name'}, inplace=True)
#print(gene_name.head(5))

df_real=pd.read_csv('zlpatch_TPM_n26.csv', index_col = 0)
df_real=df_real.drop(['c13','c19', 'c20','c22'])

index = [ 'c14','c39','c36','c38','c9','c8', 'c24', 'c23',  'c40',#INs
         'c27','c28', 'c25', 'c26','c15','c42', 'c29','c21',  'c37', 'c30', 'c41', 'c43','c16', 'c44']
index


#target_genes = ["GAD1",'GAD2','DLX1','DLX2','SLC17A6','SLC17A8','CUX2','BCL11B']
#SLC32A1 vgat
#'SLC7A7','SLC17A6','SLC17A8', vglut1 2 3
#SST EXPRESSION

#target_genes = ['SLC32A1',"GAD1",'GAD2','DLX1','DLX2','SLC7A7','SLC17A6','SLC17A8','CUX2','BCL11B','SATB2','TBR1','SLC17A5','SLC17A2','SLC18A1','SLC18A2','SLC18A3','SLC18B1']

#'GABBR1', duplicate
#target_genes = ['SST','VIP','SV2C','CALB2','CALB1',"GAD1",'GAD2','DLX1','DLX2','SYT1','RBFOX3','RORB','TLE4','CUX2','BCL11B','SATB2','TBR1','SYN1','SYN2','STX1A','NRXN1','SYP','GAP43','SV2A','SNAP25']
target_genes = ['GAD1','SATB2','SHANK1','SHANK2','SHANK3','HOMER1','SYN2','STX1A','NLGN1','NLGN2','NRXN1','SYP','SV2A','SNAP25','SYT1','SYN1','NRXN1','GAP43','DLG4','ARC','GPHN',
                'GRIA1','GRIA2','GRIA3','GRIA4','GRIN1','GRIN2A','GRIN2B','GRIN2C','GRIN2D','GRIN3A','GRIN3B',
                'GRIK1','GRIK2','GRIK3','GRIK4','GRIK5','GRM1','GRM2','GRM3','GRM4','GRM5','GRM6','GRM7','GRM8',
                'GABBR2','GABRA1','GABRA2','GABRA3','GABRA4','GABRA5','GABRA6','GABRB1','GABRB2','GABRB3','GABRD','GABRE','GABRG1',
               'GABRG2','GABRG3','GABRP','GABRQ','GABRR1','GABRR2','GABRR3',]
xyxy =gene_name[gene_name['name'].isin(target_genes)]
xy = xyxy['gene'].tolist() 
#print(xyxy)
#print(xy)
#print(type(xy))

dfplot = df_real.loc[:,xy]

dfplot.columns=xyxy['name'].tolist()
dfplot = dfplot.reindex(columns=target_genes)
dfplot=dfplot.reindex(index)
# dfplot.to_csv('./output/maypatch_More_synaptic_proteins07312020.csv')

ax,fig = plt.subplots(figsize=(20,10))

ax = sns.heatmap(dfplot, cmap="YlGnBu",robust=True) # cmap="YlGnBu" this color is color-blind friendly, and black/white printer friendly, yticklabels=False
plt.xticks(rotation=45)

# ax.figure.savefig('./output/07312020PatchseqMore_synaptic.png', dpi = 300, bbox_inches = 'tight')
plt.show()
```
