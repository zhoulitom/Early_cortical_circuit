```python
import pandas as pd
import dabest
df = pd.read_csv('patchseq_parameter2020_08.csv', index_col = 0)
#print mean and std of IN and EX
dfin = df[df['cell_type']=='IN']
dfex = df[df['cell_type']=='EX']
print('IN:\n',dfin.describe(),'\n','-_'*50)
print('EX:\n',dfex.describe())
df
```

```python
#create a new dataframe, contains wide format data
dfvrest = pd.DataFrame()
dfvrest1 = pd.DataFrame()
dfvrest2 = pd.DataFrame()
# dfvrest['EX']=
dfvrest1['EX']=df[df['cell_type']=='EX']['Cable length'][:]
dfvrest2['IN']=df[df['cell_type']=='IN']['Cable length'][:]
dfvrest = dfvrest1.append(dfvrest2, ignore_index=True)
dfvrest

two_groups = dabest.load(dfvrest, idx=("EX", "IN"),paired=False)

print(two_groups.mean_diff)

import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
two_groups.mean_diff.plot(swarm_label="Neurites cable length (um)",contrast_label="Mean differentce distribution (um)")#,raw_marker_size=10,es_marker_size=10


plt.savefig('Neurite_length_Diff_estimation.png')
```
