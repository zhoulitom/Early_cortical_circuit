```python
'''
2D grayscale	(row, col)
2D multichannel (eg. RGB)	(row, col, ch)
3D grayscale	(pln, row, col)
3D multichannel	(pln, row, col, ch)
'''
import os
filename = os.path.join( '07252019gw23_gw23_RT_day2_elecSti_1.tif')
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
#from skiamge import skimage.external.tifffile.imsave


moon = io.imread(filename)

fig, ax = plt.subplots(2,2)

ax[0,0].imshow(moon[20,:,:])
ax[0,1].imshow(moon[21,:,:])
ax[1,0].imshow(moon[27,:,:])
ax[1,1].imshow(moon[28,:,:])

fig, ax = plt.subplots()
test_minus = np.mean(np.array(moon[25:28,:,:]),axis=0)-np.mean(np.array(moon[19:23,:,:]),axis=0)
#np.mean( np.array([ old_set, new_set ]), axis=0 )
plt.imshow(test_minus)
plt.axis('off')
#io.imsave('test_minus.jpg',test_minus)
plt.imsave('./output/07252019gw23_gw23_RT_day2_elecSti_1_test_minus1.png',test_minus)

fig, ax = plt.subplots()
test_minus2 = np.mean(np.array(moon[126:129,:,:]),axis=0)-np.mean(np.array(moon[120:123,:,:]),axis=0)
#np.mean( np.array([ old_set, new_set ]), axis=0 )
plt.imshow(test_minus2)
plt.axis('off')
#io.imsave('test_minus.jpg',test_minus)
plt.imsave('./output/07252019gw23_gw23_RT_day2_elecSti_1_test_minus2.png',test_minus2)
```
